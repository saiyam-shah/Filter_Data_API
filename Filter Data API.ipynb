```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd
import time
import chromedriver_binary
from selenium.webdriver.common.action_chains import ActionChains
import sys
import random
from selenium.common.exceptions import NoSuchElementException
import json
```

```python
#Opening Chrome browser
wbD=wb.Chrome('chromedriver.exe')

#Opening webpage
wbD.get('https://www.myntra.com/men-tshirts')
#wbD.find_element(By.XPATH, "//div[@class = 'desktop-navLink']")
wbD.find_element(By.XPATH, "//div[@class = 'vertical-filters-filters categories-container']//ul[@class = 'categories-list']//li[1]").click()
time.sleep(2)
wbD.find_element(By.XPATH, "//div[@class = 'atsa-base']//ul//li[8]//div[@class = 'atsa-more']//span").click()

filter_dict = {}

wbD.find_element(By.XPATH, "//h4[text() = 'Size']").click()
key = 'Size'
try:
    wbD.find_element(By.XPATH, "//label[text() = 'M']").click()
    if key in filter_dict.keys():
        filter_dict[key].append('M')
    else:
        filter_dict[key] = ['M']
    
except NoSuchElementException:
    time.sleep(2)
    
wbD.find_element(By.XPATH, "//h4[text() = 'Pattern']").click()
key = 'Pattern'
try:
    wbD.find_element(By.XPATH, "//label[text() = 'Solid']").click()
    if key in filter_dict.keys():
        filter_dict[key].append('Solid')
    else:
        filter_dict[key] = ['Solid']
    
except NoSuchElementException:
    time.sleep(2)
    
wbD.find_element(By.XPATH, "//h4[text() = 'Sleeve Length']").click()
key = 'Sleeve Length'
try:
    wbD.find_element(By.XPATH, "//label[text() = 'Long Sleeves']").click()
    if key in filter_dict.keys():
        filter_dict[key].append('Long Sleeves')
    else:
        filter_dict[key] = ['Long Sleeves']
    
except NoSuchElementException:
    time.sleep(2)

wbD.find_element(By.XPATH, "//h4[text() = 'Sleeve Length']").click()    
wbD.find_element(By.XPATH, "//h4[text() = 'Sleeve Length']").click()
key = 'Sleeve Length'
try:
    wbD.find_element(By.XPATH, "//label[text() = 'Short Sleeves']").click()
    if key in filter_dict.keys():
        filter_dict['Sleeve Length'].append('Short Sleeves')
    else:
        filter_dict[key] = ['Short Sleeves']
        
except NoSuchElementException:
    time.sleep(2)

    
wbD.find_element(By.XPATH, "//h4[text() = 'Fit']").click()
key = 'Fit'
try:
    wbD.find_element(By.XPATH, "//label[text() = 'Slim Fit']").click()
    if key in filter_dict.keys():
        filter_dict[key].append('Slim Fit')
    else:
        filter_dict[key] = ['Slim Fit']
        
except NoSuchElementException:
    time.sleep(2)


wbD.find_element(By.XPATH, "//div[@class = 'vertical-filters-filters brand-container']//div[@class = 'brand-more']").click()

brand_list = []
l = random.sample(range(2,80),5)
print(l)
for i in l:
    location_a="//div[@class = 'FilterDirectory-panel FilterDirectory-expanded']//div//ul[@class = 'FilterDirectory-list']//li["
    location_b="]"
    location = location_a + str(i) + location_b
    try:
        wbD.find_element(By.XPATH, location).click()
        time.sleep(2)
        location_c = location + "//label"

        brand_name = wbD.find_element(By.XPATH, location_c).text
        for i in range(len(brand_name)):
            if brand_name[i] == '(':
                brand_name = brand_name[0:i]
                break
        print(brand_name)
        brand_list.append(brand_name)
    except NoSuchElementException:
        location_a="//div[@class = 'FilterDirectory-panel FilterDirectory-expanded']//div//ul[@class = 'FilterDirectory-list']//li["
        location_b="]"
        location = location_a + str(i+1) + location_b
        wbD.find_element(By.XPATH, location).click()
        time.sleep(2)
        location_c = location + "//label"
        brand_name = wbD.find_element(By.XPATH, location_c).text
        for i in range(len(brand_name)):
            if brand_name[i] == '(':
                brand_name = brand_name[0:i]
                break
        print(brand_name)
        brand_list.append(brand_name)

print(brand_list)
    
wbD.find_element(By.XPATH, "//div[@class = 'FilterDirectory-panel FilterDirectory-expanded']//div//span").click()

time.sleep(5)
wbD.close()
```

```python
myntra_dict = filter_dict
myntra_dict['Brands'] = brand_list

with open('myntra_api.json', 'w') as json_file:
    json.dump(myntra_dict, json_file)
```
